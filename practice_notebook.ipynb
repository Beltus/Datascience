## Normalize Death Columns in Pandas

```python
#
from sklearn import preprocessing
#create a new column X, to hold the float values of the column to be normalized

X = dataset[['Deaths']].values.astype(float)

#create a maximum and minimum preprocessor object
min_max_scaler = preprocessing.MinMaxScaler()

#create an object to transform the data to fit min max processor
#X.reshape(1 , -1)

X_scaled = min_max_scaler.fit_transform(X)

#convert the normalize object back to a dataframe

norm_X = pd.DataFrame(X_scaled)

norm_X.columns = ['norm']

#norm_X.rename(columns = {'0' : 'norm'} , inplace = True)

norm_X.head()
```

```python
norm_X['norm'].plot.hist(1000)
```
